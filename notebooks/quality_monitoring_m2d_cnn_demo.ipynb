# AI-Enabled Quality Monitoring Demo

This notebook demonstrates how to work with the summary data for the m2D-CNN control chart pattern recognition project.

It covers:

- Model accuracy comparison
- Model stability comparison
- Simulated control chart pattern examples


```
import pandas as pd
import matplotlib.pyplot as plt

models = pd.read_csv("../data/model_comparison.csv")
patterns = pd.read_csv("../data/control_chart_patterns.csv")
sample_patterns = pd.read_csv("../data/sample_control_chart_patterns.csv")

models
```

## Accuracy comparison

```
plt.figure(figsize=(10, 5))
plt.bar(models["model"], models["accuracy_percent"])
plt.title("Model accuracy comparison")
plt.ylabel("Accuracy (%)")
plt.ylim(95, 100)
plt.xticks(rotation=20)
plt.grid(axis="y", linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

## Standard deviation comparison

```
plt.figure(figsize=(10, 5))
plt.bar(models["model"], models["std_deviation"])
plt.title("Model stability comparison")
plt.ylabel("Standard deviation")
plt.xticks(rotation=20)
plt.grid(axis="y", linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

## Simulated control chart patterns

```
plt.figure(figsize=(12, 6))

offset = 0
for pattern_name, group in sample_patterns.groupby("pattern_name"):
    plt.plot(group["time_index"], group["value"] + offset, label=pattern_name)
    offset += 7

plt.title("Examples of simulated control chart patterns")
plt.xlabel("Time index")
plt.ylabel("Signal with vertical offsets")
plt.grid(True, linestyle="--", alpha=0.4)
plt.legend(ncol=3, bbox_to_anchor=(0.5, -0.15), loc="upper center")
plt.tight_layout()
plt.show()
```

## Key interpretation

The m2D-CNN model improves recognition accuracy by combining visual process behavior from control charts and histograms with numerical statistics such as mean, standard deviation, and skewness.
